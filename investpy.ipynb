```python
def indicator(symbol,country,interval):
    import investpy

    search_result = investpy.search_quotes(text=symbol, products=['stocks'],
                                       countries=[country], n_results=1)
    technical_indicators = search_result.retrieve_technical_indicators(interval=interval)
    print(technical_indicators)
```

```python
indicator('NTPC','INDIA','daily')
```
